```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
import plotly.express as pe
%matplotlib inline
```

```python
data = pd.read_json('./data/proc/data.json', orient='split')
```

```python
category = data.groupby(['date', 'category']).agg({'cost': 'sum'}).reset_index()
```

```python
pe.line(category, x='date', y='cost', color='category', width=1000, template='plotly_white', title="Daily cost")
```

```python
rolling_mean = category.groupby('category').rolling(30, on='date').sum().reset_index()
pe.line(rolling_mean, x='date', y='cost', color='category', width=1000, template='plotly_white', title="30 day average cost")
```
